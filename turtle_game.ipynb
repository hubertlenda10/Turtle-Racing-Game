**<p style="font-size:16px" >As my turtle project I decided to build a simple race game between two turtles.</font></p>**
**<p style="font-size:16px" >Before turtles start race, they have to create a background and a road.</font></p>**
**<p style="font-size:16px" >Who first cross finish line is a winner.</font></p>**
<br>

<p>To build the game I used for loops which helped me to reduce code during repetitive turtle activities while creating a background and a road, and during the race, to allow turtles move. Use of if statements helped me to check which of the turtles first crossed finish line and what distance travelled. Distances of both are printed on the bottom of the file.</p>



```python
# Import all necessary moduls
import time
import turtle
from turtle import Turtle
from random import randint

# Window setup
window = turtle.Screen()
window.title =("Racing game")
turtle.bgcolor("lime green")
turtle.color("white")
turtle.speed(0)
turtle.penup()

# Set position of turtle
turtle.setpos(-300,-350)

# Set turtle to create a road
turtle.color("red")
turtle.pensize(10)
turtle.fillcolor("light slate gray")
turtle.begin_fill()
turtle.pendown()
turtle.left(90)
for i in range(2):
    turtle.forward(750)
    turtle.right(90)
    turtle.forward(310)
    turtle.right(90)
turtle.end_fill()

# Set the name of the race
turtle.penup()
turtle.setpos(-285,265)
turtle.write("TURTLE RACE",font = ("Arial",30,"bold"))
turtle.penup()


# Set current position of turtle and set size of pen
turtle.setpos(-150,-370)
turtle.pensize(10)
turtle.pendown()

# For loop to create multiple lines
for i in range(8):
    turtle.forward(70)
    turtle.penup()
    turtle.forward(70)
    turtle.pendown()
    
turtle.penup()

# Create first row of squares    
turtle.setpos(-295,240)
turtle.pensize(1)
turtle.pendown()
turtle.right(90)


for i in range(5):
    turtle.color("white")
    turtle.fillcolor("white")
    turtle.begin_fill()
    for i in range(4):
        turtle.forward(30)
        turtle.right(90)
    turtle.forward(30)
    turtle.end_fill()

    turtle.color("black")
    turtle.fillcolor("black")
    turtle.begin_fill()
    for i in range(4):
        turtle.forward(30)
        turtle.right(90)
    turtle.forward(30)
    turtle.end_fill()
turtle.penup()


# Create second raw of squares
turtle.setpos(-295,210)
for i in range(5):
    turtle.color("black")
    turtle.fillcolor("black")
    turtle.begin_fill()
    for i in range(4):
        turtle.forward(30)
        turtle.right(90)
    turtle.forward(30)
    turtle.end_fill()

    
    turtle.color("white")
    turtle.fillcolor("white")
    turtle.begin_fill()
    for i in range(4):
        turtle.forward(30)
        turtle.right(90)
    turtle.forward(30)
    turtle.end_fill()
turtle.hideturtle()


# Create tree
turtle.goto(180,100)
turtle.left(90)
turtle.fillcolor("brown")
turtle.begin_fill()

# Create bottom of tree
for i in range(2):
    turtle.forward(100)
    turtle.right(90)
    turtle.forward(10)
    turtle.right(90)
turtle.end_fill()
turtle.penup()
turtle.forward(100)
turtle.right(90)
turtle.forward(5)

# Create top of the tree
turtle.fillcolor("green")
turtle.begin_fill()
turtle.circle(60)
turtle.end_fill()


# Create top of the tree
turtle.fillcolor("green")
turtle.begin_fill()
turtle.circle(40)
turtle.end_fill()


# Creating turtle1
turtle1 = Turtle()
turtle1.speed(0)
turtle1.color("pink")
turtle1.shape("turtle")
turtle1.penup()
turtle1.goto(-225,-160)
turtle1.left(90)

# Creating turtle2
turtle2 = Turtle()
turtle2.speed(0)
turtle2.color("yellow")
turtle2.shape("turtle")
turtle2.penup()
turtle2.goto(-75,-160)
turtle2.left(90)

# Pause the game for a second before starting a race
time.sleep(1)


for i in range(137):
    turtle1.forward(randint(1,4))
    turtle2.forward(randint(1,4))

# Create variables to store current position of the turtles
x = turtle1.ycor()
y = turtle2.ycor()

# Positions of the turtles on the finish line
turtle1.pos()
print("Position of first: ")
print(round(turtle1.ycor()))

turtle2.pos()
print("Position of second: ")
print(round(turtle2.ycor()))


# Create if statement to check which turtle has traveled the longer way

if x > y:
    turtle1.left(360)
    turtle1.write("--WINNER")
else:
    turtle2.left(360)
    turtle2.write("--WINNER")

    
    
turtle.exitonclick()


turtle.done()
```
